```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('UrWay_Expanded_Dataset1.csv')
df_2 = pd.read_csv('UrWay_Expanded_Dataset2.csv')
df_3 = pd.read_csv('UrWay_Expanded_Dataset3.csv')
df_4 = pd.read_csv('UrWay_Expanded_Dataset4.csv')
df_5 = pd.read_csv('UrWay_Expanded_Dataset5.csv')
df_6 = pd.read_csv('UrWay_Deep_Dataset6.csv')
df_7 = pd.read_csv('UrWay_Deep_Dataset7.csv')
```

```python
print("Dataset 1 shape:", df_1.shape)
print("Dataset 2 shape:", df_2.shape)
print("Dataset 3 shape:", df_3.shape)
print("Dataset 4 shape:", df_4.shape)
print("Dataset 5 shape:", df_5.shape)
print("Dataset 6 shape:", df_6.shape)
print("Dataset 7 shape:", df_7.shape)
```

```python
df_1 = df_1.sample(frac=1, random_state=42).reset_index(drop=True)
df_2 = df_2.sample(frac=1, random_state=42).reset_index(drop=True)
df_3 = df_3.sample(frac=1, random_state=42).reset_index(drop=True)
df_4 = df_4.sample(frac=1, random_state=42).reset_index(drop=True)
df_5 = df_5.sample(frac=1, random_state=42).reset_index(drop=True)
df_6 = df_6.sample(frac=1, random_state=42).reset_index(drop=True)
df_7 = df_7.sample(frac=1, random_state=42).reset_index(drop=True)
print("Shuffling completed.")
```

```python
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7], ignore_index=True)
print("Combined dataset shape:", df.shape)
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
print("Final shuffling completed.")
```

```python
df.sample(20)
```

```python
df['title'] = df['title'].astype(str) + " " + df['SubTopic'].fillna('').astype(str)
df = df.drop(['SubTopic'], axis=1)
```

```python
df.sample(20)
```

```python
df.to_csv('UrWay Final Dataset.csv', index=False)
print("Final dataset saved as 'UrWay Final Dataset.csv'.")
```
