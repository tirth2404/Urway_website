```python
import pandas as pd

# Read the three mental wellness CSVs (FIXED PATHS — only test file is in repo)
# NOTE: mental_wellness_train.csv and mental_wellness_tracker.csv are NOT in the repo.
# Only mental_wellness_test.csv is present under archive (3)/.
# Add the missing files to ml/DataSet/archive (3)/ before running the full merge.

df01 = pd.read_csv("../DataSet/archive (3)/mental_wellness_train.csv")
df02 = pd.read_csv("../DataSet/archive (3)/mental_wellness_tracker.csv")
df03 = pd.read_csv("../DataSet/archive (3)/mental_wellness_test.csv")
```

```python
print(df01.shape)
print(df02.shape)
print(df03.shape)
```

```python
df0 = pd.concat([df01, df02, df03], axis=0)
df0.shape
```

```python
df0 = df0.drop_duplicates()
df0.shape
```

```python
df0.head()
```

```python
df = df0.drop(columns=[
    'User_ID', 'Date', 'Mood_Score', 'Screen_Time_Hours',
    'Social_Interaction_Hours', 'Work_Productivity_Score', 'Weather'
])

df.head()
```

```python
# Save to Mental Wellness folder (FIXED PATH)
out_path = "../Physical Information/Mental Wellness/mental_wellness.csv"
df.to_csv(out_path, index=False)
print(f"Saved to {out_path}")
```
