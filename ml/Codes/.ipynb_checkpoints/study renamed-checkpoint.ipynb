```python
import pandas as pd

# Read the raw numbered-column dataset
# NOTE: The original file is NOT present in ml/DataSet/ in the current repo.
# You must add the raw CSV (originally named 'DATA (1).csv') to ml/DataSet/ before running.
# This notebook renames its numeric columns to human-readable names and writes Data_renamed.csv.
df0 = pd.read_csv("../DataSet/Data_renamed.csv")
```

```python
df0.head()
```

```python
column_names = {
    1:  "Student Age",
    2:  "Sex",
    3:  "Graduated High School Type",
    4:  "Scholarship Type",
    5:  "Additional Work",
    6:  "Regular Artistic or Sports Activity",
    7:  "Has Partner",
    8:  "Total Salary",
    9:  "Transportation to University",
    10: "Accommodation Type",
    11: "Mother Education",
    12: "Father Education",
    13: "Number of Siblings",
    14: "Parental Status",
    15: "Mother Occupation",
    16: "Father Occupation",
    17: "Weekly Study Hours",
    18: "Reading Frequency (Non-Scientific)",
    19: "Reading Frequency (Scientific)",
    20: "Seminars / Conferences Attendance",
    21: "Impact of Projects on Success",
    22: "Class Attendance",
    23: "Midterm Preparation (With Whom)",
    24: "Midterm Preparation (When)",
    25: "Taking Notes in Class",
    26: "Listening in Class",
    27: "Discussion Improves Success",
    28: "Flip Classroom Usefulness",
    29: "Last Semester CGPA",
    30: "Expected Graduation CGPA"
}

# Rename using column positions
df0.columns = [
    column_names.get(i + 1, col)
    for i, col in enumerate(df0.columns)
]

df0.head()
```

```python
# Save the renamed dataset back to DataSet folder (FIXED PATH)
out_path = "../DataSet/Data_renamed.csv"
df0.to_csv(out_path, index=False)
print(f"Saved to {out_path}")
```
